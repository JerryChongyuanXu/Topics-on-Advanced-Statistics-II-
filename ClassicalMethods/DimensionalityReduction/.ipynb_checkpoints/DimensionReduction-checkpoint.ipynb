```python
import numpy as np
import matplotlib.pyplot as plt
import os
import struct
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import MinMaxScaler
import math
```

```python
def load_mnist(path, kind = 'train'):
    #Load MNIST data from 'path'
    labels_path = os.path.join(path, '%s-labels-idx1-ubyte' % kind)
    images_path = os.path.join(path, '%s-images-idx3-ubyte' % kind)
    
    #The with statement gurantees that the file will be closed no matter how the nested block exits
    #Open arguments: 'rb' -> r=reading b=binary file mode
    #Refer to the opened file as libpath
    with open(labels_path, 'rb') as libpath:
        #(>) big-endian and II refers to 2 usigned integers of 4 bytes each
        #we read in 8 bytes and stores them as the magic number and the number of items
        #See the MNIST website for the data description
        magic, n = struct.unpack('>II',libpath.read(8))
        #fromfile is a function in the numpy package
        #A highly efficient way of reading binary data with a known data-type, 
        #as well as parsing simply formatted text files. 
        #Since count was not supplied it will read in all values
        labels = np.fromfile(libpath, dtype = np.uint8)
        #similar but reading in 4 unsigned integers to start
    with open(images_path, 'rb') as imgpath:
        magic, num, rows, cols = struct.unpack(">IIII",imgpath.read(16))
        #the reshape option transforms the array into the correct type of matrix
        #nx784-- note the tru data is a 28x28 image
        images = np.fromfile(imgpath, dtype = np.uint8).reshape(num,rows*cols)
    print('Observations labels: %d, Observations Images: %d' % (n,num))
    return images, labels
```

```python
def print_mnist_Embedding(X,labels,title,location,labels_dict):
        markers = iter(['<','s','o','^','*','v','1','2','3','4'])
        colors = np.array(['b','g','r','c','m','y','k','gray','orange','darkgreen'])
        n = np.unique(labels)
        for i in n:
            plt.scatter(x=X[labels == i,0], y = X[labels == i,1],color = colors[i],marker = next(markers))
        plt.title(title)
        #plt.legend(list(labels_dict.values()),loc ='best')
        plt.xticks([], [])
        plt.yticks([], [])
        art = []
        lgd = plt.legend(list(labels_dict.values()),loc=9, bbox_to_anchor=(1.1, 1.0))
        art.append(lgd)
        plt.savefig(location ,dpi=900,additional_artists=art,bbox_inches="tight")
        plt.show()
```

```python
mnist_loc = "../Data"
```

```python
X_train, y_train = load_mnist(mnist_loc,kind = "train")
rescale = MinMaxScaler(feature_range=(-1, 1))
X_train = rescale.fit_transform(X_train)
```

```python
X_test, y_test = load_mnist(mnist_loc,kind = "t10k")
rescale = MinMaxScaler(feature_range=(-1, 1))
X_test = rescale.fit_transform(X_test)
```

```python
labels_dict = {0:'Zero'
,1:'One'
,2:'Two'
,3:'Three'
,4:'Four'
,5:'Five'
,6:'Six'
,7:'Seven'
,8:'Eight'
,9:'Nine'}
```

## t-Distributed Stochastic Neighbor Embedding

```python
tsne_Train = TSNE(n_components=2).fit_transform(X_train)
tsne_Test = TSNE(n_components=2).fit_transform(X_test)
```

```python
location = "Images/Embed_t-SNE_test"
title = "t-SNE embedding for MNIST (Test Set)"
print_mnist_Embedding(tsne_Test,y_test,title,location,labels_dict)
np.savetxt(mnist_loc + "/MNIST_2D_Test.csv",tsne_Test)
np.savetxt(mnist_loc + "/MNIST_Test_Labels.csv",y_test,fmt ='%d')
```

```python
location = "Images/Embed_t-SNE_train"
title = "t-SNE embedding for MNIST (Train Set)"
print_mnist_Embedding(tsne_Train,y_train,title,location,labels_dict)
np.savetxt(mnist_loc + "/MNIST_2D_Train.csv",tsne_Train)
np.savetxt(mnist_loc + "/MNIST_Train_Labels.csv",y_train,fmt ='%d')
```

# PCA

```python
def principal(data,p):
    #Extract eigen vectors and values
    ev, eig = np.linalg.eigh(np.cov(data,rowvar=False))
    #Keep the top p-components
    idx = ev.argsort()[::-1]
    eig = eig[:,idx][:,0:p]
    ev = ev
    top_ev = ev[idx][0:p]
    if p != data.shape[1]:
        print("Top {} princiapl components explains {:.2f}% of the variance".format(p,100*np.sum(top_ev)/np.sum(ev)))
        if p>2:
            print_mnist_principal(eig.T)
    return eig, top_ev

def print_mnist_principal(images):
    obs= np.arange(images.shape[0])
    n = math.ceil(math.sqrt(images.shape[0]))
    if n> 5:
        n = 5
    fix,ax = plt.subplots(nrows = n, ncols = n,sharex = True, sharey = True,)
    ax = ax.flatten()
    for i in obs:
        if i >= n*n:
            break
        img = images[i,:].reshape(28,28)
        ax[i].imshow(img,cmap = 'Greys')
    ax[0].set_xticks([])
    ax[0].set_yticks([])
    plt.tight_layout()
    plt.show()
    return

def plot_totalVariation(eigen_vals,stop = 25):
    #Graph amount of variation explained by eigen top prinicipal components
    tot= np.sum(eigen_vals)
    var_exp = sorted(eigen_vals/tot,reverse = True)
    cum_var_exp = np.cumsum(var_exp)
    n = np.min([eigen_vals.shape[0],stop])
    plt.bar(np.arange(n),var_exp[0:stop],alpha = 0.5, align = 'center',label = 'Individual explained variance')
    plt.step(np.arange(n), cum_var_exp[0:stop],where ='mid',label = 'Cumulative explained variance')
    plt.title('Total Variation Explained by the top {} components.'.format(n))
    plt.ylabel('Explained variance ratio')
    plt.xlabel('Principal components')
    plt.legend(loc = 'best')
    plt.show()
```

```python
eig_vec, eig_val = principal(X_train,X_train.shape[1])
```

```python
plot_totalVariation(eig_val,25)
```

```python
eig_vec, eig_val = principal(X_train,25)
```

```python
eig_vec, eig_val = principal(X_train,2)
X_pca = np.matmul(X_train,eig_vec)
```

```python
location = "Images/Embed_PCA_train"
title = "PCA for MNIST (Train Set)"
print_mnist_Embedding(X_pca,y_train,title,location,labels_dict)
np.savetxt(mnist_loc + "/MNIST_2D_PCA_Train.csv",X_pca)
```

```python
eig_vec, eig_val = principal(X_test,X_test.shape[1])
```
